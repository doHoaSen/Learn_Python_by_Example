```python
#129
#텍스트 박스에 숫자 입력하도록 요청, variable.isdigit() 이용해 정수 입력했는지 확인하는 버튼 만들기
#정수라면 목록에 추가, 그렇지 않으면 엔트리 박스의 내용 지우기
#목록의 내용 지우는 버튼도
from tkinter import *

def add_num():
    num = num_box.get()
    if num.isdigit():
        num_list.insert(END, num)
        num_box.delete(0,END)
        num_box.focus()
    else:
        num_box.delete(0,END)
        num_box.focus()
        
        
def clear_list():
    num_list.delete(0, END)
    num_box.focus()
    
    
    
window = Tk()
window.title("Number list")
window.geometry("400x200")

label1 = Label(text = "Enter a number: ")
label1.place(x = 20, y = 20, width = 100, height = 25)

num_box = Entry(text = "")
num_box.place(x = 120, y = 20, width = 100, height = 25)
num_box.focus()

btn1 = Button(text = "Add to list", command = add_num)
btn1.place(x = 250, y = 20, width = 100, height = 25)

num_list = Listbox()
num_list.place(x = 120, y = 50, width = 100, height = 100)

btn2 = Button(text = "Clear list", command = clear_list)
btn2.place(x = 250, y = 50, width = 100, height = 25)

window.mainloop()
```

```python
#130
from tkinter import *

def add_num():
    num = num_box.get()
    if num.isdigit():
        num_list.insert(END, num)
        num_box.delete(0,END)
        num_box.focus()
    else:
        num_box.delete(0,END)
        num_box.focus()
        
        
def clear_list():
    num_list.delete(0, END)
    num_box.focus()
    
    
def save_list():
    f = open("numbers.csv", "w")
    tmp_list = num_list.get(0, END)
    for x in tmp_list:
        newrecord = x + "\n"
        f.write(str(newrecord))
    f.close()
    
    
window = Tk()
window.title("Number list")
window.geometry("400x200")

label1 = Label(text = "Enter a number: ")
label1.place(x = 20, y = 20, width = 100, height = 25)

num_box = Entry(text = "")
num_box.place(x = 120, y = 20, width = 100, height = 25)
num_box.focus()

btn1 = Button(text = "Add to list", command = add_num)
btn1.place(x = 250, y = 20, width = 100, height = 25)

num_list = Listbox()
num_list.place(x = 120, y = 50, width = 100, height = 100)

btn2 = Button(text = "Clear list", command = clear_list)
btn2.place(x = 250, y = 50, width = 100, height = 25)

btn3 = Button(text = "Save list", command = save_list)
btn3.place(x = 250, y = 80, width = 100, height = 25)

window.mainloop()
```
