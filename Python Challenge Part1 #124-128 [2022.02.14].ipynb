# Tkinter GUI

**GUI**: Graphical User Interface

==> 사용자가 더욱 사용자 친화적인 방식으로 프로그램 사용할 수 있도록 프로그래머에게 화면, 텍스트 박스, 버튼 등 생성할 수 있도록 기능 제공

**Tkinter**은 이를 가능하게 해주는 **라이브러리**

```python
from tkinter import *

def Call():
    msg = Label(window, text = "You pressed the button")    #메시지 표시하는 텍스트를 화면에 추가
    msg.place(x = 30, y = 50)
    button["bg"] = "blue"    #버튼의 배경색
    button["fg"] = "white"   #메시지 박스의 폰트(포그라운드) 색상 지정
    

window = Tk()
window.geometry("200x110")#윈도우의 크기 결정
button = Button(text = "Press me", command = Call)
button.place(x = 30, y = 20, width = 120, height = 25)    #윈도우 안에서 각 항목의 위치 결정
window.mainloop()
```

# Tkinter 

- entry_box = Entry(text = 0) : 비어있는 엔트리 박스 생성. 엔트리 박스는 사용자로부터 데이터 입력받거나 결과 표시 위해 사용


- output_box = Message(text = 0) : 결과를 표시하기 위해 사용되는 메시지박스 생성


- output_box["relief"] = "sunken" : 메시지 박스의 스타일 지정 

    -->flat, raised, sunken, grooved, ridged


- list_box = listbox() : 문자열만 포함할 수 있는 드롭다운 리스트 박스 생성


- entry_box["justify"] = "center" : 엔트리 박스의 텍스트 행 정렬 지정 (메시지 박스에는 작동되지 않음)


- button1 = Button(text = "Click here", command = click) : click이라는 함수 실행하는 버튼 생성


- label.place(x = 50, y = 20, width = 100, height = 25) : 윈도우에 표시될 객체 위치 지정


- entry_box.delete(0, END) : 엔트리 박스(또는 리스트) 내용 삭제


- num = entry_box.get() : 엔트리 박스의 내용을 num이라는 변수에 저장 (메시지 박스에서는 작동되지 않음)


- answer = output_box["text"] : 메시지 박스의 내용을 answer이라는 변수에 저장 (엔트리 박스에서는 작동하지 않음)


- output_box["text"] = total : 메시지 박스의 내용을 total이라는 변수의 값으로 변경함


- window.mainloop() : 프로그램을 계속 작동하게 하기 위하여 이 코드는 프로그램의 하단에 있어야 함

```python
#126
#사용자가 숫자 입력하도록 요청하는 프로그램 생성
#사용자가 버튼 클릭하면 입력한 숫자를 합계에 더하게 되며, 그 합이 화면에 출력됨
#이 과정 원하는 만큼 반복
#합계 0으로 돌리고 표시했던 합계를 원래 상태로 비워 다시 시작할 준비하는 버튼도 있어야 함
# sum 함수, reset 함수 필요

from tkinter import *


def add():
    num = enter_txt.get()
    num = int(num)
    answer = output_txt["text"]
    answer = int(answer)
    total = num + answer
    output_txt["text"] = total

def reset():
    output_txt["text"] = 0
    enter_txt.delete(0, END)
    enter_txt.focus()


window = Tk()
window.title("Adding on and on")
window.geometry("450x100")

#입력
enter_lbl = Label(text = "Enter a number: ")
enter_lbl.place(x = 50, y = 20, width = 100, height = 25)

enter_txt = Entry(text = 0)
enter_txt.place(x = 150, y = 20, width = 100, height = 25)
enter_txt.focus()

#버튼
add_btn = Button(text = "Add", command = add)
add_btn.place(x = 300, y = 20, width = 50, height = 25)

reset_btn = Button(text = "Reset", command = reset)
reset_btn.place(x = 300, y = 50, width = 100, height = 25)


#결과
output_lbl = Label(text = "Output is ")
output_lbl.place(x = 50, y = 50, width = 100, height = 25)

output_txt = Message(text = 0)
output_txt.place(x = 150, y = 50, width = 100, height = 25)
output_txt["bg"] = "white"
output_txt["relief"] = "sunken"

window.mainloop()
```

```python
#127
#텍스트박스에 이름 입력하도록 요청하는 윈도우 생성
#입력한 이름을 목록 하단에 추가하는 버튼 생성
#목록의 내용 지우는 버튼 생성

from tkinter import *

def add_name():
    name = name_txt.get()
    name_list.insert(END, name)
    name_txt.delete(0, END)
    name_txt.focus()
    
    
    
def clear_list():
    name_list.delete(0, END)
    name_txt.focus()

    
window = Tk()
window.title("Name List")
window.geometry("400x200")

name_lbl = Label(text = "Enter a Name: ")
name_lbl.place(x = 20, y = 20, width = 100, height = 25)

name_txt = Entry(text = 0)
name_txt.place(x = 120, y = 20, width = 100, height = 25)

add_btn = Button(text = "Add to list", command = add_name)
add_btn.place(x = 250, y = 20, width = 100, height = 25)

name_list = Listbox()
name_list.place(x = 120, y = 50, width = 100, height = 100)

del_btn = Button(text = 'Clear list', command = clear_list)
del_btn.place(x = 250, y = 50, width = 100, height = 25)

window.mainloop()
```

```python
#128
#키로미터와 마일 변환하는 프로그램 만들기

from tkinter import *

def mtok():
    mile = input_txt.get()
    mile = float(mile)
    message = mile * 1.6093
    output.delete(0, END)
    output.insert(END, message)
    output.insert(END, "km")


def ktom():
    km = input_txt.get()
    km = float(km)
    message = km * 0.6214
    output.delete(0, END)
    output.insert(END, message)
    output.insert(END, "mile")



window = Tk()
window.title("Convert km to mile")
window.geometry("260x200")

#입력
input_lbl = Label(text = "Enter the value you want to convert: ")
input_lbl.place(x = 15, y = 20)

input_txt = Entry(text = "")
input_txt.place(x = 30, y = 50, width = 200, height = 25)
input_txt["justify"] = "center"
input_txt.focus()


#버튼
convert1 = Button(text = "Convert mile into km", command = mtok)
convert1.place(x=  30, y = 80, width = 200, height = 25)

convert2 = Button(text = "Convert km to mile", command = ktom)
convert2.place(x=  30, y = 110, width = 200, height = 25)

output = Entry(text = "")
output.place(x = 30, y = 140, width = 200, height = 25)
output["justify"] = "center"

window.mainloop()
```
