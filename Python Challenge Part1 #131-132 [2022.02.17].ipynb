```python
#131
#새로운 csv 만드는 프로그램 만들기
#이름과 나이 입력하도록 요청, 입력한 데이터를 방금 만든 파일에 추가하도록
from tkinter import *
import csv

def create_new():
    f = open("ages.csv", "w")
    f.close()
    
def save_list():
    f = open("ages.csv", "a")
    name = namebox.get()
    age = agebox.get()
    newrecord = name + ", " + age + "\n"
    f.write(str(newrecord))
    f.close()
    namebox.delete(0, END)
    agebox.delete(0, END)
    namebox.focus()
    
    
window = Tk()
window.title("People list")
window.geometry("400x100")

lbl1 = Label(text = "Enter a name: ")
lbl1.place(x = 20, y = 20, width = 100, height = 25)

namebox = Entry(text = "")
namebox.place(x = 120, y = 20, width = 100, height = 25)
namebox["justify"] = "left"
namebox.focus()

lbl2 = Label(text = "Enter their age: ")
lbl2.place(x = 20, y = 50, width = 100, height = 25)

agebox = Entry(text = "")
agebox.place(x = 120, y = 50, width = 100, height = 25)
agebox["justify"] = "left"

btn1 = Button(text = "Create new file", command = create_new)
btn1.place(x = 250, y = 20, width = 100, height = 25)

btn2 = Button(text = "Add to file", command = save_list)
btn2.place(x = 250, y = 50, width = 100, height = 25)

window.mainloop()
```

```python
#132
#131번에서 만든 파일의 내용 가져와 목록에 표시되도록 사람들의 이름과 나이 목록에 추가하는 프로그램 만들기

from tkinter import *
import csv

    
def save_list():
    f = open("ages.csv", "a")
    name = namebox.get()
    age = agebox.get()
    newrecord = name + ", " + age + "\n"
    f.write(str(newrecord))
    f.close()
    namebox.delete(0, END)
    agebox.delete(0, END)
    namebox.focus()
    

def read_list():
    name_list.delete(0, END)
    f = list(csv.reader(open("ages.csv")))
    tmp = []
    for row in f:
        tmp.append(row)
    for data in tmp:
        name_list.insert(END, data)
        
        
window = Tk()
window.title("People list")
window.geometry("400x200")

lbl1 = Label(text = "Enter a name: ")
lbl1.place(x = 20, y = 20, width = 100, height = 25)

namebox = Entry(text = "")
namebox.place(x = 120, y = 20, width = 100, height = 25)
namebox["justify"] = "left"
namebox.focus()

lbl2 = Label(text = "Enter their age: ")
lbl2.place(x = 20, y = 50, width = 100, height = 25)

agebox = Entry(text = "")
agebox.place(x = 120, y = 50, width = 100, height = 25)
agebox["justify"] = "left"

btn1 = Button(text = "Create new file", command = save_list)
btn1.place(x = 250, y = 20, width = 100, height = 25)

btn2 = Button(text = "Add to file", command = read_list)
btn2.place(x = 250, y = 50, width = 100, height = 25)

lbl3 = Label(text = 'Saved names: ')
lbl3.place(x = 250, y = 80, width = 100, height = 25)

name_list = Listbox()
name_list.place(x = 120, y = 80, width = 230, height = 100)
window.mainloop()
```
