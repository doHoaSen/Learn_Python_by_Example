```python
#119
#낮은 숫자와 높은 숫자를 입력하도록 요청하는 함수 정의, 두 값 사이의 임의의 숫자를 생성하여 comp_num에 저장
#메시지를 출력하고 사용자가 생각하는 숫자 입력하라고 요청하는 함수 하나 더 정의
#사용자가 입력한 숫자와 comp_num 일치하는지 확인하는 함수 .. 맞을 때까지  계속 반복
import random

def enternum():
    lownum = int(input("Enter a low number: "))
    highnum = int(input("Enter a high number: "))
    comp_num = random.randint(lownum, highnum)
    return comp_num
    

def enterusernum():
    print("I am thinking of a number: ")
    usernum = int(input("Enter a number: "))
    return usernum
    

def sameornot(user, comp):#n1: usernum, n2:comp_num
    notsame = True
    while notsame:
        if user == comp:
            print("Correct, you win")
            notsame = False
        elif user > comp:
            user = int(input("Too high. Try again: "))
        else:
            user = int(input("Too low. Try again: "))
                
#main
comp_num = enternum()
usernum = enterusernum()
sameornot(usernum, comp_num)
```

```python
#120
import random


#def addition
def addition():
    n1 = random.randint(5, 20)
    n2 = random.randint(5, 20)
    ans = n1 + n2
    print(n1, "+", n2, "=", "?")
    userans = int(input("What is the answer? : "))
    data = (ans, userans)
    return data

#def subtraction
def subtraction():
    n1 = random.randint(25, 50)
    n2 = random.randint(1, 25)
    ans = n1 - n2
    print(n1, "-", n2, "=", "?")
    userans = int(input("What is the answer? : "))
    data = (ans, userans)
    return data

#def checkans
def checkans(data):
    if data[0] == data[1]:
        print("Correct")
    else:
        print("Incorrect, the answer is ", data[0])

#main
print("1) Addition \n2) Subtraction")
user = int(input("Enter 1 or 2 : "))
if user == 1:
    data = addition()
elif user == 2:
    data = subtraction()
    
checkans(data)
```

```python
#121
#이름 목록 관리하는 프로그램 생성

def addname():
    name = input("Enter a name: ")
    names.append(name)
    return names

def changename():
    num = 0
    for x in names:
        print(num, x)
        num += 1
    selectnum = int(input("Enter a number of the name you want to change: "))
    name = input("enter new name: ")
    names[selectnum] = name
    return names

def delname():
    num = 0
    for x in names:
        print(num, x)
        num += 1
    selectnum = int(input("Enter a number of the name you want to delete: "))
    del names[selectnum]
    return names

def shownames():
    for x in names:
        print(x)
    print()


again = True
names = []
while again:
    print("1) Add a Name \n2) Change a name \n3) Delete a name \n4) View names \n5) Quit")
    select = int(input("What do you want to do? : "))
    if select == 1:
        names = addname()
    elif select == 2:
        names = changename()
    elif select == 3:
        names = delname()
    elif select == 4:
        names = shownames()
    elif select == 5:
        again = False
    else:
        print("Incorrect option")
```

```python
#123
def addtofile():
    f = open('Salaries.csv', 'a')
    name = input("Enter a name: ")
    salary = int(input("Enter a salary: "))
    newrecord = name + ", " + str(salary) + "\n"
    f.write(str(newrecord))
    f.close()

def showrecords():
    f = open('Salaries.csv', 'r')
    for row in f:
        print(row)
    f.close()
    
def delrecord():
    f = open('Salaries.csv', 'r')
    tmp = []
    for row in f:
        tmp.append(row)
    f.close()
    num = 0
    for row in tmp:
        print(num, row)
        num += 1
    delrow = int(input("Enter a row number to delete: "))
    del tmp[delrow]
    
    f = open('Salaries.csv', 'w')
    for row in tmp:
        f.write(row)
    f.close()
    

#main
again = True
while again:
    print("""1) Add to file \n2) View all records \n3) Delete a record \n4) Quit program""")
    select = int(input("Enter a number you want to do: "))
    if select == 1:
        addtofile()
    elif select == 2:
        showrecords()
    elif select == 3:
        delrecord()
    elif select == 4:
        again = False
    else:
        print("Wrong number. Try again.")
```
